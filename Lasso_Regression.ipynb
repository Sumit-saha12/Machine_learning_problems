```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
```

```python
x,y = make_regression(n_features=1, n_samples=10000,noise=9,random_state=1)
```

```python
plt.scatter(x,y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

### Taking Bydefault Alpha value

```python
model = Lasso()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model.coef_
```

```python
pred=model.predict(x)
plt.scatter(x,y)
plt.plot(x,pred,c='r')
```

### Taking Alpha=10

```python
model=Lasso(alpha=10)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model.coef_
```

```python
pred = model.predict(x)
plt.scatter(x,y)
plt.plot(x,pred,c='r')
```

### Taking Alpha=50

```python
model=Lasso(alpha=50)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model.coef_
```

```python
pred=model.predict(x)
plt.scatter(x,y)
plt.plot(x,pred,c='r')
```

### Taking alpha=1000

```python
model=Lasso(alpha=1000)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model.coef_
```

```python
pred = model.predict(x)
plt.scatter(x,y)
plt.plot(x,pred,c='r')
```
