```python
from sklearn.datasets import make_regression
from sklearn.linear_model import Ridge,Lasso
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
x,y = make_regression(n_samples=10000,noise=10,n_features=1,random_state=1)
```

```python
plt.scatter(x,y,s=3)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

### Taking bydefault Alpha value

```python
model = Ridge()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model.coef_
```

```python
pred = model.predict(x)
```

```python
plt.scatter(x,y,label='actual',s=3)
plt.plot(x,pred,label='predicted',c='r')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

### Taking Alpha=100

```python
model = Ridge(alpha=100)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model.coef_
```

```python
pred = model.predict(x)
plt.scatter(x,y,label='acutal',s=3)
plt.plot(x,pred,label='predicted',c='r')
plt.legend()
plt.show()
```

### Taking Alpha=10000

```python
model = Ridge(alpha=10000)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model.coef_
```

```python
pred=model.predict(x)
plt.scatter(x,y,label='acutal',s=3)
plt.plot(x,pred,label='predicted',c='r')
plt.legend()
plt.show()
```

### Taking Alpha=10000000

```python
model = Ridge(alpha=100000000)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model.coef_
```

```python
pred = model.predict(x)
plt.scatter(x,y,label='acutal',s=3)
plt.plot(x,pred,label='predicted',c='r')
plt.legend()
plt.show()
```

## Show here Ridge regularization shows that when alpha value increasing, value of co-eff is decreasing and accuracy is also compromise. When i take alpha = 1 then predicted line is best fit line but when i take alpha=1000000 then predicted line is horizontal.
## In Ridge regularization, Co-eff value is not exact zero. It's going to zero but not go to exact zero.
