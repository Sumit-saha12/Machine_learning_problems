```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import pandas as pd
```

```python
data = pd.read_csv('encoding.csv')
data.head(3)
```

## Label Encoder

```python
enc = LabelEncoder()
```

```python
data1=data.copy()
```

```python
data1['TEAM']=enc.fit_transform(data1['TEAM'])
data1
```

## One Hot Encoding

```python
ohenc = OneHotEncoder()
```

```python
data2=data.copy()
# data2['TEAM']=data2['TEAM'].reshape(-1,1)
```

```python
f=pd.DataFrame(ohenc.fit_transform(data2[['TEAM']]).toarray())
f
```

```python
data2=data2.join(f)
```
