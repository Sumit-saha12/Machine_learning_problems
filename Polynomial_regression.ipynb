```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
x=np.array([2,3,4,5,6,7,8,9])
y=np.array([4,8,3,6,5,1,10,5])
```

```python
plt.scatter(x,y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train=x_train.reshape(-1,1)
x_test=x_test.reshape(-1,1)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
x=x.reshape(-1,1)
```

```python
pred = model.predict(x)
plt.scatter(x,y,label='acutal')
plt.plot(x,pred,label='predicted')
plt.legend()
plt.show()
```

```python
poly= PolynomialFeatures(degree=2)
poly_x=poly.fit_transform(x)
poly_x
```

```python
model = LinearRegression()
model.fit(poly_x,y)
model.score(poly_x,y)
```

```python
pred=model.predict(poly_x)
plt.scatter(x,y)
plt.plot(x,pred)
```

```python
for i in range(1,10):
    poly=PolynomialFeatures(degree=i)       ## Take Polynomial Features
    poly_x=poly.fit_transform(x)            ## Transform X using Poly
    model=LinearRegression()                ## Take Linear Regression as a Model
    model.fit(poly_x,y)                     ## Fit data into Model
    pred=model.predict(poly_x)              ## Model predict
    print(f'Degree: {i}')                   ## Print Degree
    plt.scatter(x,y,label='Actual')         ## Plot data points
    plt.plot(x,pred,label='Predict',c='r')  ## Plot Predictate line
    plt.legend()                  
    plt.show()
    if model.score(poly_x,y)>=0.99:         ## If model score is greater than 99%, Execution will be closed
        break
```
