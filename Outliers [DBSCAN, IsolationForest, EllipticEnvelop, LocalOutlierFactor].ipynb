```python
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
from sklearn.ensemble import IsolationForest
from sklearn.covariance import EllipticEnvelope
import matplotlib.pyplot as plt
import numpy as np
from sklearn.neighbors import LocalOutlierFactor
```

```python
x,y = make_blobs(n_samples=500,centers=3,cluster_std=0.4,random_state=0)
plt.scatter(x[:,0],x[:,1],c=y)
```

## DBSCAN

```python
model = DBSCAN(eps=0.3, min_samples=5).fit(x)
```

```python
plt.scatter(x[:,0],x[:,1],c=model.labels_)
```

## Isolation Forest

```python
iff = IsolationForest()
iff.fit(x)
```

```python
pred=iff.predict(x)
plt.scatter(x[:,0],x[:,1],c=pred)
```

## Elliptic Envelop

```python
x,y = make_blobs(n_samples=500,centers=2,cluster_std=6,random_state=1)
plt.scatter(x[:,0],x[:,1])
```

```python
ee = EllipticEnvelope(contamination=0.1)
ee.fit(x)
```

```python
pred=ee.predict(x)
plt.scatter(x[:,0],x[:,1],c=pred)
```

## Local Outlier Function

```python
x,y = make_blobs(n_samples=500,centers=3,cluster_std=0.4,random_state=0)
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
lof = LocalOutlierFactor(n_neighbors=3)
lof.fit(x)
```

```python
pred=lof.fit_predict(x)
plt.scatter(x[:,0],x[:,1],c=pred)
```

```python
dp = np.array([[1,5],[1,6],[1,7],[2,5],[2,7],[5,4],[10,8]])
plt.scatter(dp[:,0],dp[:,1])
```

```python
x,y=dp[:,0],dp[:,1]
```

```python
x=x.reshape(-1,1)
```

```python
lof = LocalOutlierFactor(n_neighbors=3)
lof.fit(x)
```

```python
pred = lof.fit_predict(x)
```

```python
plt.scatter(x,y,c=pred)
```
