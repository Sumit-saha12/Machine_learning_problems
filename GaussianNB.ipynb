```python
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
```

```python
x=np.array([80,70,75,65,45,67,88,90,76,73,34])
y=np.array([0,0,0,1,1,1,0,0,1,1,0])
```

```python
plt.scatter(x,y,c=y)
```

```python
x=x.reshape(-1,1)
model = GaussianNB()
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
x,y=make_classification(n_features=5,n_classes=2,n_samples=10000,random_state=50)
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
model=GaussianNB()
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
pred = model.predict(x)
```

```python
matrix = confusion_matrix(pred,y)
matrix
```

```python
matrix_display= ConfusionMatrixDisplay(matrix,display_labels=model.classes_)
matrix_display.plot()
```
