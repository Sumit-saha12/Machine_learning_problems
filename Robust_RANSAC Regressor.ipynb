```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, RANSACRegressor
```

```python
x,y = make_regression(n_samples=1000000,random_state=1,noise=5,n_features=1)
```

### Added outlier in below step

```python
outliers=300000
y[:outliers]=600+np.random.normal(size=outliers)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
lr = LinearRegression()
rr = RANSACRegressor()
```

```python
lr.fit(x_train,y_train)
rr.fit(x_train,y_train)
```

```python
print(lr.score(x_test,y_test))
print(rr.score(x_test,y_test))
```

```python
lr_pred=lr.predict(x_test)
rr_pred=rr.predict(x_test)
```

```python
plt.scatter(x_test,y_test,label='actual_value',s=1,color='violet')
plt.plot(x_test,lr_pred,label='Linear_regrssion',c='r')
plt.plot(x_test,rr_pred,label='RANSAC_regression',color='green')
plt.legend()
plt.show()
```
